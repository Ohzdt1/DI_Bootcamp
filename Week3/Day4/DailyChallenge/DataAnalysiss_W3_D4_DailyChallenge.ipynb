```python
import matplotlib.pyplot as plt

# Scatter plot to explore the relationship between Social Support and Happiness Score
plt.figure(figsize=(10, 6))
plt.scatter(df['Social support'], df['Score'], alpha=0.9, c='purple', edgecolors="green", s=100)

# chart
plt.xlabel("Social Support")
plt.ylabel("Happiness Score")
plt.title("Relationship Between Social Support and Happiness Score (2019)")

plt.show()
```

```python
import seaborn as sns

# Create a subplot with 2 axes: one for GDP and one for Healthy Life Expectancy
fig, ax1 = plt.subplots(figsize=(40, 10))

# Bar plot for GDP per Capita
sns.barplot(x=df['Country or region'], y=df['GDP per capita'], ax=ax1, color='purple', label="GDP per Capita")
ax1.set_xlabel("Country or Region")
ax1.set_ylabel("GDP per Capita")
ax1.set_title("GDP per Capita and Healthy Life Expectancy across Countries (2019)")
ax1.tick_params(axis='x', rotation=90)  # Rotate x-axis labels for readability

# Line plot for Healthy Life Expectancy
ax2 = ax1.twinx()  # Create a second y-axis that shares the same x-axis
sns.lineplot(x=df['Country or region'], y=df['Healthy life expectancy'], ax=ax2, color='green', marker='o', label="Healthy Life Expectancy")

ax2.set_ylabel("Healthy Life Expectancy")
ax2.legend(loc='upper right')

plt.tight_layout()  # Adjust layout for better fit
plt.show()
```
