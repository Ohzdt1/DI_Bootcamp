# **ðŸŒŸ Exercise 2: Creating a Line Plot for Temperature Variation**

```python
import matplotlib.pyplot as plt

# Define data
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
temperatures = [72, 74, 76, 80, 82, 78, 75]

print(days)
print(temperatures)
```

```python
# Create the plot
plt.plot(days, temperatures, marker='o', linestyle='-', color='b', label="Temperature")

plt.xlabel("Day")
plt.ylabel("Temperature (Â°F)")
plt.title("Temperature Variation Over a Week")
plt.legend()

plt.show()
```

# **ðŸŒŸ Exercise 3: Visualizing Monthly Sales with a Bar Chart**

```python
import matplotlib.pyplot as plt
```

```python
# Define data
months = ["January", "February", "March", "April", "May"]
sales = [5000, 5500, 6200, 7000, 7500]

print(months)
print(sales)
```

```python
# Create the bar chart
plt.bar(months, sales, color='green')

plt.xlabel("Month")
plt.ylabel("Sales Amount ($)")
plt.title("Monthly Sales Data")

plt.show()
```

# **ðŸŒŸ Exercise 4: Visualizing the Distribution of CGPA with a Histogram**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Sample CGPA data
data = {"CGPA": [2.5, 3.0, 3.2, 3.5, 3.7, 3.8, 3.9, 4.0, 2.8, 3.6, 3.3, 2.9, 3.4, 3.1, 3.0]}
df = pd.DataFrame(data)

print(data)
print(df)
```

```python
# Create the histogram
sns.histplot(df["CGPA"], bins=14, kde=True, color='purple')

# Add labels and title
plt.xlabel("CGPA")
plt.ylabel("Frequency")
plt.title("Distribution of CGPA")

# Display the plot
plt.show()
```

# **ðŸŒŸ Exercise 5: Comparing Anxiety Levels Across Different Genders**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Sample data
data = {
    "Gender": ["Male", "Female", "Female", "Male", "Male", "Female", "Female", "Male", "Male", "Female"],
    "Anxiety": ["Yes", "No", "Yes", "Yes", "No", "No", "Yes", "Yes", "Yes", "No"]
}

# Create DataFrame
df = pd.DataFrame(data)

print(df)
```

```python
# Create the bar plot
sns.countplot(x="Gender", hue="Anxiety", data=df, palette="Set1")

# Add labels and title
plt.xlabel("Gender")
plt.ylabel("Count of Students")
plt.title("Anxiety Levels Across Different Genders")

# Display the plot
plt.show()
```

# **ðŸŒŸ Exercise 6: Exploring the Relationship Between Age and Panic Attacks**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Sample data
data = {
    "Age": [18, 20, 22, 19, 21, 23, 18, 25, 24, 22],
    "Panic_Attacks": ["Yes", "No", "Yes", "No", "Yes", "No", "Yes", "Yes", "No", "Yes"]
}

# Create DataFrame
df = pd.DataFrame(data)

print(df)
```

```python
# Convert Panic_Attacks to numeric values
df["Panic_Attacks"] = df["Panic_Attacks"].map({"Yes": 1, "No": 0})

print(df)
```

```python
# Create the scatter plot
sns.scatterplot(x="Age", y="Panic_Attacks", data=df, hue="Panic_Attacks", palette="bright", s=100)

# Add labels and title
plt.xlabel("Age")
plt.ylabel("Panic Attacks (1 = Yes, 0 = No)")
plt.title("Relationship Between Age and Panic Attacks")

# Display the plot
plt.show()
```
