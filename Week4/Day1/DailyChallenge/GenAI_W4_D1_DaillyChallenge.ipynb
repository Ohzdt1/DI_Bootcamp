```python
temp_data = np.random.uniform(low=-5, high=35, size=(10, 12))

temp_data = np.array(temp_data).astype(int)

print(temp_data)
```

```python
cities = ['City ' + str(i+1) for i in range(10)]

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

df = pd.DataFrame(temp_data, index=cities, columns=months)

print(df)
```

```python
annual_avgtemps = df.mean(axis=1).astype(int)

print(annual_avgtemps)
```

```python
city_highest_temp = annual_avgtemps.idxmax()
city_lowest_temp = annual_avgtemps.idxmin()

print(df)
print(f"\nCity with the highest average temp: {city_highest_temp} ({annual_avgtemps[city_highest_temp]}°C)")
print(f"City with the lowest average temp: {city_lowest_temp} ({annual_avgtemps[city_lowest_temp]}°C)")

```

```python
plt.figure(figsize=(10, 6))

for city in df.index:
    plt.plot(df.columns, df.loc[city], marker='o', linestyle='-', label=city, linewidth=3)

# Title and labels
plt.title('Monthly Temp Trends for Each City', fontsize=14, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)', fontsize=14, fontweight='bold')
plt.legend(title='Cities', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df, annot=True, fmt="d", cmap="coolwarm", linewidths=1, cbar=True)

plt.title('Temperature Trends Across Cities')
plt.xlabel('Month')
plt.ylabel('City')

plt.yticks(rotation=0)


plt.show()
```
